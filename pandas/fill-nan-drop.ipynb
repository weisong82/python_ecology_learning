# Filling with a PandasObject

```python
dff = pd.DataFrame(np.random.randn(7,3), columns=list('ABC'))
dff.iloc[3:5,0] = np.nan
dff.iloc[4:6,1] = np.nan
dff.iloc[5:8,2] = np.nan
dff
```

```python
dff.fillna(dff.mean())
```

```python
dff1=dff.fillna(dff.mean()['B':'C'])
dff1
```
