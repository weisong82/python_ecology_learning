```python
#DataFrame
dates = pd.date_range('20130101',periods=6)
dates

df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=list('ABCD'))
df
```

```python
####看一个colum的全部取值
df.A.unique()
```

# 对一列的值进行改造   map映射， lambda函数


```python
df['E'] = df.A.astype('str').map(str.lower)

df['F'] = df.B.apply(lambda x: abs(x))
df
```

# lambda + ifelse 条件计算

```python
df['N']=df['A'].apply(lambda x: True if  x>0 else False) 
df
```

# Assign 赋值函数

```python
df.assign(Z= df.A+df.B).head()
```

# combine 合并

```python
df1 = pd.DataFrame({'A' : [1., np.nan, 3., 5., np.nan],
                        'B' : [np.nan, 2., 3., np.nan, 6.]})
df1   

df2 = pd.DataFrame({'A' : [5., 2., 4., np.nan, 3., 7.],
                      'B' : [np.nan, np.nan, 3., 4., 6., 8.]})
df2
#Combining overlapping data sets  重叠df聚合
#  df1优先，有值的地方取df1值，nan的地方才补df2的值
df1.combine_first(df2) 
```

```python
#General DataFrame Combine  一般性合并  名称相同的列；接受一个合并函数
# passes the combiner function pairs of Series (i.e., columns whose names are the same).
combiner = lambda x, y: np.where(pd.isna(x), y, x) ## 当x项未na的时候，填充y

df1.combine(df2, combiner)
```
